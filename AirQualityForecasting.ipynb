## Data Collection & Preprocessing

```python
import numpy as np
import pandas as pd
df_airquality = pd.read_csv("/content/AirQualityUCI.csv" ,sep = ";" ,decimal = ",")
df_airquality.head()
```

```python
# Remove last 2 columns that have null values
df_airquality = df_airquality.iloc[: ,:-2]
df_airquality.head()
```

```python
df_airquality.tail()
```

```python
df_airquality.shape
```

```python
df_airquality = df_airquality.dropna(how = "all")
```

```python
df_airquality.tail()

#the rows which have null values are deleted
```

```python
df_airquality.shape
```

```python
#getting information about data

df_airquality.info()
```

```python
# getting the sum of missing of missing values in all columns

df_airquality.isnull().sum()
```

```python
# null values are represnted by "-200 "
# getting the count of -200

df_airquality[df_airquality == -200].count()
```

```python
#replacing -200 with NAN values
df_airquality = df_airquality.replace(to_replace= -200 , value = np.nan)
df_airquality.tail()
```

```python
df_airquality.isnull().sum()
```

```python
df_airquality.info()
```

```python
#converting date into YYYY-MM-DD
df_airquality["Date"] = pd.to_datetime(df_airquality["Date"] ,format="%d/%m/%Y")
df_airquality["Time"] = pd.to_datetime(df_airquality["Time"] ,format="%H.%M.%S").dt.time
df_airquality["DateTime"] = df_airquality.apply(lambda x: pd.to_datetime(x["Date"]).replace(hour=x["Time"].hour, minute=x["Time"].minute, second=x["Time"].second), axis=1)
df_airquality.head()
```

```python
df_airquality.drop(["Date" , "Time"] , axis = 1 , inplace = True)
df_airquality.head()
```

```python
df_airquality.info()
```

```python
#finding the mean of all columns
df_airquality.mean()
```

```python
#replacing NAN values with the mean
df_airquality.fillna(df_airquality.mean() , inplace = True)
```

```python
df_airquality.tail()
```

# FB Prophet Model

```python
df_airquality.info()
```

```python
#creating a new dataframe 'data'
data = pd.DataFrame({
    "ds" : df_airquality["DateTime"],
    "y" : df_airquality["RH"]
})
```

```python
data.head()
```

```python
from prophet import Prophet
```

# Model Training

```python
model = Prophet()
model.fit(data)
```

```python
#making future predictions
future = model.make_future_dataframe(periods=365, freq = "H")
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

```

```python
fig1 = model.plot(forecast)

```

```python
fig2 = model.plot_components(forecast)
```
